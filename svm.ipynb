```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([[8, 3], [3, 4], [-4, 7]])
y = np.array([1, 1, -1])
```

```python
w = np.linalg.solve(np.c_[x, np.ones(x.shape[0])], 1 / y)
w
```

```python
xx, yy = np.meshgrid(
    np.linspace(-5, 10, 1000),
    np.linspace(-5, 10, 1000),
)
X = np.c_[xx.ravel(), yy.ravel()]
X = np.c_[X, np.ones(X.shape[0])]
```

```python
ypred = (X @ w > 0).astype(int)

fig, ax = plt.subplots(figsize=(8, 8))
ax.grid()
ax.axis([-5, 10, -5, 10])
ax.set_aspect(1)
ax.contourf(xx, yy, ypred.reshape(xx.shape), alpha=0.5, cmap=plt.cm.Blues)
ax.scatter(*x.T, c=y, s=200)
```
